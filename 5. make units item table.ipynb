```python
match_df = pd.read_csv('C:/Users/user/Documents/GitHub/Final-Project-A09-/game_results_of_each_participants.csv')
```

```python
match_df = match_df.drop(columns=['augments','traits'])
```

#### 게임 종료시에 유닛이 가진 아이템 테이블 만들기

##### 각 참가자의 최종 조합 데이터를 챔피언 이름으로 각각의 열로 분리하기

```python
import ast

# 'units' 열의 문자열을 파이썬 리스트로 변환하여 가공
match_df['units'] = match_df['units'].apply(ast.literal_eval)
processed_df = match_df.explode('units').reset_index(drop=True)

# 가공된 데이터프레임 출력
processed_df
```

##### units 행을 세개의 행으로 분리하고 데이터를 각각 담기

```python
import pandas as pd

# units 열을 분할하여 새로운 열을 생성하는 함수 정의
def extract_unit_info(row):
    unit_info = row['units']
    if isinstance(unit_info, dict):
        return pd.Series({
            'character_id': unit_info['character_id'],
            'itemNames': unit_info['itemNames'],
            'tiers': unit_info['tier']  # tier만 추출하여 tiers 생성
        })
    else:
        return pd.Series({
            'character_id': None,
            'itemNames': None,
            'tiers': None
        })

# 분할된 열을 포함한 새로운 데이터프레임 생성
processed_df[['character_id', 'itemNames', 'tiers']] = processed_df.apply(extract_unit_info, axis=1)

# 기존 units 열 삭제
processed_df.drop('units', axis=1, inplace=True)

# 결과 출력
processed_df
```

```python
processed_df = processed_df[processed_df['character_id'].notnull()]

processed_df.info()
```

```python
# csv 파일로 저장
processed_df.to_csv('units_item_list.csv', index=False)
```
