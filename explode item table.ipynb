##### itemNames 컬럼이 그냥 문자열의 형태로 되어있으므로 리스트 형태로 바꿔주어야함

```python
item_table['itemNames'] = item_table['itemNames'].apply(eval)
```

##### 챔피언에 들어가있는 아이템 3개가 모두 동일하더라도 순서가 다를수 있으므로 리스트안의 아이템들을 먼저 정렬해줘야함

```python
item_table['itemNames'] = item_table['itemNames'].apply(sorted)

# 확인을 위해 특정 챔피언의 아이템이 정렬되었는지 확인
bard_df = item_table[item_table['챔피언'] == '바드']

bard_df.head()
```

##### item1,2,3 컬럼을 생성한 뒤에 리스트 안의 아이템을 순서대로 값으로 넣어줌

```python
# itemNames 열의 값을 이용하여 item1, item2, item3 열에 값 할당
# 리스트 안의 갯수가 특정 갯수보다 적으면 값이 없는 item1,2,3 컬럼은 null 이 되도록함
item_table['item1'] = item_table['itemNames'].apply(lambda x: x[0] if len(x) >= 1 else None)
item_table['item2'] = item_table['itemNames'].apply(lambda x: x[1] if len(x) >= 2 else None)
item_table['item3'] = item_table['itemNames'].apply(lambda x: x[2] if len(x) >= 3 else None)
```

##### itemNames 컬럼을 문자열로 바꾸어준 itemNames 컬럼을 만들어 추가(통계 편의상)

```python
item_table['itemNames_str'] = item_table['itemNames'].apply(lambda x: ', '.join(x))
```

##### 바드 아이템만 가지고 EDA 구하기 테스트

```python
# 바드 df 정의
bard_df = item_table[item_table['챔피언'] == '바드']

bard_df.info()
```

```python
# 'itemNames_str'을 기준으로 그룹화한 후 'placement', 'last_round', 'tiers' 컬럼의 평균값 구하기
group_means = bard_df.groupby('itemNames_str')[['placement', 'last_round', 'tiers']].mean()

# 각 그룹의 데이터 수(count) 구하기
group_counts = bard_df.groupby('itemNames_str').size()

# 그룹 내의 데이터 수와 함께 정렬
result_df = pd.concat([group_means, group_counts], axis=1)
result_df.columns = ['placement_mean', 'last_round_mean', 'tiers_mean', 'data_count']
result_df = result_df.sort_values(by='data_count', ascending=False)

result_df
```
