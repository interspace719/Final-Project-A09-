```python
import pandas as pd
import json
from pandas import json_normalize
import numpy as np

# 최대 출력 컬럼 제한 해제
pd.set_option('display.max_colwidth', None)
```

```python
item_tier_table = pd.read_csv('챔피언-장착 아이템 테이블.csv')

# 필요한 컬럼만 남기기
item_tier_table =  item_tier_table[['placement' , '챔피언', '아이템1', '아이템2', '아이템3']]

item_tier_table.head()
```

```python
# '아이템1', '아이템2', '아이템3' 컬럼을 '아이템' 열로 변환하고, NaN 값은 제외합니다.
item_tier_table = pd.melt(item_tier_table, id_vars=['placement', '챔피언'], value_vars=['아이템1', '아이템2', '아이템3'], value_name='아이템').dropna()

# 원하는 컬럼 드랍하기
item_tier_table.drop(columns=['variable', '챔피언'], inplace=True)

print(item_tier_table)
```

##### 아이템 선택 확률 컬럼 추가

```python
# 아이템별 등장 횟수를 계산하여 item_pickrate 열을 추가합니다.
item_tier_table['item_pickrate'] = item_tier_table.groupby('아이템')['아이템'].transform('count') / len(item_tier_table) * 100

print(item_tier_table)
```

##### 아이템 이름으로 그룹바이, pickrate 와 placement 는 평균으로 테이블 생성

```python
item_tier_table = item_tier_table.groupby('아이템').mean().reset_index().round(3)

item_tier_table.sort_values(by='item_pickrate', ascending=False)
```

```python
item_tier_table['아이템'].unique()
```

##### 통계값이 필요없는 아이템 제거(재료 아이템, 케일 아이템 제거)

```python
# 만든 데이터 테이블에서 재료 아이템 행 삭제
item_tier_table = item_tier_table[~item_tier_table['아이템'].isin(['곡궁', '뒤집개', '여신의 눈물', 'B.F. 대검',
                                                                '쓸데없이 큰 지팡이', '거인의 허리띠', '쇠사슬 조끼',
                                                                 '음전자 망토', '연습용 장갑' ])]

# 문신 아이템 행 + 전력가의 왕관 행 삭제
item_tier_table = item_tier_table[~item_tier_table['아이템'].isin(['독의 문신', '보호의 문신', '분노의 문신', '폭격의 문신',
                                                                '활력의 문신', '힘의 문신', '전략가의 왕관'])]

# 문신 아이템 행 + 전력가의 왕관 행 삭제
item_tier_table = item_tier_table[~item_tier_table['아이템'].isin(['속도의 부적', '지원의 부적', '무력의 부적', 
                                                                '가속의 두루마리', '힘의 두루마리',
                                                                '능력의 고서', '민첩의 고서', '치유의 고서'])]


item_tier_table.rename(columns={'placement': 'avg_placement'}, inplace=True)

item_tier_table.sort_values(by='item_pickrate', ascending=False)
```

##### 아이템 제작 레시피 테이블과 left join

```python
item_recipe = pd.read_csv('일반 아이템 제작 레시피.csv')

item_recipe.rename(columns={'아이템명': '아이템'}, inplace=True)

item_recipe
```

```python
# 아이템 컬럼을 기준으로 left 조인합니다.
merged_df = pd.merge(item_tier_table, item_recipe, on='아이템', how='left')

print(merged_df.sort_values(by='item_pickrate', ascending=False))
```

```python
# 중복된 행을 제거합니다.
merged_df = merged_df.drop_duplicates()

merged_df.sort_values(by='item_pickrate', ascending=False).head(15) ## 123 개 확인
```

```python
merged_df.to_csv('아이템 티어 테이블_평균 등수 + 픽률(태블로용 최종)).csv', index=False)
```
